```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time 
```

```python
data = pd.read_csv("/Users/florianlichius/Daten/bluebikes_2017_2018")
data.head()
data = pd.DataFrame(data)
data.head()
```

```python
#data.drop("number", axis=1, inplace=True)
test = pd.DataFrame(data['bike_id'])
test
```

```python
# Number of observations per bike
plt.hist(data['bike_id'],bins=100)
plt.xlabel('Bike ID')
plt.ylabel('Number of Observations')
plt.title('Number of Observations per Bike')
plt.show()
```

```python
# Number of observations per start station
fig, ax = plt.subplots(figsize=(15, 7))
plt.hist(data['start_station_id'], bins=100)
plt.xlabel('Station ID')
plt.ylabel('Number of Observations')
plt.title('Number of Observations per Station')
plt.show()
```

```python
# Number of observations per end station
fig, ax = plt.subplots(figsize=(15, 7))
plt.hist(data['end_station_id'], bins=100)
plt.xlabel('Station ID')
plt.ylabel('Number of Observations')
plt.title('Number of Observations per Station')
plt.show()
```

```python

fig, ax = plt.subplots(figsize=(15,7))
stations = [data['start_station_id'], data['end_station_id']]
ax.hist(stations, bins=100, label=['Start Station', 'End Station'] )
plt.xlabel('Station ID')
plt.ylabel('Number of Observations')
plt.title('Number of Observations per Station')
plt.legend(loc='upper right')
plt.show()
```
