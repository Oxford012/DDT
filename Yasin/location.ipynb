```python

#funtion that takes the path to the csv file and returns a folium map with a random datasets locations out of massachusetts_loc
def getMap(pathToCSV):

    ma = pd.read_csv(pathToCSV)

    # random row drom massachusetts_pois
    ma.sample(1)
    
    massachusetts_pois = gpd.GeoDataFrame(ma, geometry=gpd.points_from_xy(ma['lon'], ma['lat']))
    massachusetts_map = folium.Map(location=(ma.sample(1).lat , ma.sample(1).lon),tiles='OpenStreetMap',zoom_start=12, control_scale=True, max_zoom=20)
    for i in range(0,len(massachusetts_pois)):
        folium.Marker([massachusetts_pois.iloc[i]['lat'], massachusetts_pois.iloc[i]['lon']], popup=massachusetts_pois.iloc[i]['name']).add_to(massachusetts_map)
    return massachusetts_map

```

```python

def plotHeatMap(data, map, radius, blur):
    data = data[['lat', 'lon']]
    data = [[row['lat'],row['lon']] for index, row in data.iterrows()]
    HeatMap(data, radius=radius, blur=blur).add_to(map)
    return map
```
